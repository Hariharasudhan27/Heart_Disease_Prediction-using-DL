```python
import pandas as pd
import numpy as np
import tensorflow as tf

# Load your dataset
# Replace 'your_data.csv' with the path to your dataset file
data = pd.read_csv('/content/drive/MyDrive/HCA mini & Ass 2 dataset/Heart_assignment_modified.csv')

# Define the target variable and predictor variables
X = data[['Age', 'RestBP', 'Chol']]
y = data['MaxHR']

# Build an ANN model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=(X.shape[1],)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)  # Output layer with 1 neuron for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, y, epochs=100, batch_size=32, verbose=1)  # Train on the entire dataset

# Accept user input for Age, RestBP, and Chol
user_age = float(input("Enter Age: "))
user_restbp = float(input("Enter Resting Blood Pressure (RestBP): "))
user_chol = float(input("Enter Cholesterol (Chol): "))

# Prepare user input for prediction
user_input = np.array([[user_age, user_restbp, user_chol]])

# Make predictions on the user input
user_maxhr = model.predict(user_input)[0, 0]

# Print the predicted MaxHR for the user input
print(f"Predicted MaxHR: {user_maxhr}")
```
